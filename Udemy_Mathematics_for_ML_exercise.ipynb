```python
def regression(my_x,my_m,my_b):
  return my_x*my_m + my_b
```

```python
def regression_plot(my_x,my_m,my_b,my_y):
  fig , ax = plt.subplots()
  ax.scatter(my_x,my_y)
  x_min, x_max = ax.get_xlim()
  y_min, y_max = regression(x_min,my_m,my_b).detach().item(), regression(x_max,my_m,my_b).detach().item()
  ax.set_xlim([x_min,x_max])
  a = ax.plot([x_min,x_max],[y_min,y_max])
```

 1. $x=3, y = 0$
 1. $x=2,y=3$
 1. $x=-2, y=-3$

###### **Step 1** : Get the graph of the curve in the x-z axis

```python
def f_aut(my_x,my_y):
  return my_x**2 - my_y**2
xs = np.linspace(-5,5,1000)
z = f_aut(xs,0)
fig, ax = plt.subplots()
plt.xlabel ('X-axis')
plt.ylabel ('Z-axis')
plt.title ('f(x,y) = x^2 - y^2')
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
a = ax.plot(xs,z)
```

###### **Step 2** : Setting up pytorch

Setting up the x and y points

```python
x_pts = [3., 2., -2.]
y_pts = [0., 3., -3.]
```

Defining in the tangent line function

```python
def x_tgt(x_vals,my_x, my_y,fn,colour):
  x = torch.tensor(my_x, requires_grad=True)
  y = torch.tensor(my_y, requires_grad=False)
  z = fn(x,y)
  plt.scatter(my_x,z.detach().item(), c=colour, zorder=3)
  z.backward()
  #getting the gradient as the slope
  slp = x.grad
  #b = z - mx
  intcpt = z.detach().item() - slp*my_x
  x_tgt = slp*x_vals + intcpt
  plt.plot(x_vals,x_tgt,c=colour, linestyle= 'dashed', linewidth = 0.7, zorder = 3)
```

Calling the function created above in a function call

```python
fig, ax = plt.subplots()
plt.xlabel ('X-axis')
plt.ylabel ('Z-axis')
plt.title ('X-Z Tangents')
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.ylim(-1,20)
a = ax.plot(xs,z)
colour = ['red','blue','green']
for i, x_val in enumerate(x_pts):
  x_tgt(xs,x_val,0,f_aut,colour[i])
```

###### **Step 3** : Repeat the steps for y-z axis

Defining the _y tangent_ function

```python
def y_tgt(y_vals,my_x, my_y,fn,colour):
  x = torch.tensor(my_x, requires_grad=False)
  y = torch.tensor(my_y, requires_grad=True)
  z = fn(x,y)
  plt.scatter(my_y,z.detach().item(), c=colour, zorder=3)
  z.backward()
  #getting the gradient as the slope
  slp = y.grad
  #b = z - mx
  intcpt = z.detach().item() - slp*my_y
  y_tgt = slp*y_vals + intcpt
  plt.plot(y_vals,y_tgt,c=colour, linestyle= 'dashed', linewidth = 0.7, zorder = 3)
```

Defining ys variable and creating the plot

```python
ys = np.linspace(-5,5,1000)
z = f_aut(0,ys)
fig, ax = plt.subplots()
plt.xlabel ('Y-axis')
plt.ylabel ('Z-axis')
plt.title ('f(x,y) = x^2 - y^2')
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.ylim(-12,1)
a = ax.plot(ys,z)
colour = ['red','blue','green']
for i, y_val in enumerate(y_pts):
  y_tgt(ys,0,y_val,f_aut,colour[i])
```

###### **Solution for Problem 1**

```python
x = [48, 50, 54, 60, 49, 55, 59, 62]
def p_from_t(ds,dslen,baseline):
  # finding the t value
  t = (np.mean(ds)-baseline)/st.sem(ds)
  #returning the t statistic
  return 2 * st.t.cdf(-abs(t),dslen-1)
print (f't-statistic is {p_from_t(x,len(x),50)}')
if p_from_t(x,len(x),50) < 0.05:
  print('Significant difference with GMO yeast')
else:
  print('No significant difference with GMO yeast')
```

#### **Chapter 3 : Linear Regression**

Create OLS regression to predict body of Adele Penguin from flipper length

```python
#getting the body mass and flipper length in separate numpy vectors
all_adelie = sns.load_dataset('penguins')
adelie_1 = all_adelie[all_adelie['species']=='Adelie']
body_mass = adelie_1[adelie_1.sex=='Female']['body_mass_g'].to_numpy()/1000
flipper = adelie_1[adelie_1.sex=='Female']['flipper_length_mm'].to_numpy()
```

Doing a scatterplot of the data.

```python
sns.set_style('darkgrid')
_ = sns.scatterplot(x=flipper,y=body_mass)
plt.xlabel('Flipper length (mm)')
plt.ylabel('Body mass (kg)')
```

Now let us estimate $\beta_1$ using the formula
$$\hat{\beta}_1 = \frac{\text{cov}(x,y)}{σ^2_x}$$

```python
cov_p = np.cov(flipper,body_mass)[0,1]
var_p = np.var(flipper)
beta_1 = cov_p/var_p
print(f'beta_1 is {beta_1}')
cov_mat = np.cov(flipper,body_mass)
print (f'beta_1 using cov_mat is {cov_mat[0,1]/cov_mat[0,0]}')
```

Personal inference from above is that there is a different in the way variance is calculated in the covariance matrix and using np.var in numpy

```python
mean_mass = body_mass.mean()
mean_flipper = flipper.mean()
```

Let us calculate $\hat{\beta_0}$ using the equation

$$ \hat{\beta_0} = \bar{y} - \hat{\beta_1}\bar{x} $$

```python
beta_0 = mean_mass - beta_1*mean_flipper
print(f'beta_0 is {beta_0}')
```

Let us use the above data to predict the mass of a penguin with flipper length **197 mm**

```python
x_i = 197
y_i = beta_0 + beta_1*x_i
print(f'Predicted body mass is {y_i} kilograms')
```

Fit the line and also provide the point in the line in the scatterplot

```python
flipper_line = np.linspace(170,205,100)
mass_line = beta_0 + beta_1*flipper_line
_ = sns.scatterplot(x=flipper,y=body_mass)
plt.plot(flipper_line,mass_line,color='orange')
_= plt.scatter(x_i,y_i,color='purple')
```

```python
y = titanic.survived
clas = pd.get_dummies(titanic['class'], dtype = 'int64')
gender = pd.get_dummies(titanic['sex'], dtype = 'int64')
X = pd.concat([clas.First, clas.Second, gender.male ,gender.female,titanic.age],axis=1)
X = sm.add_constant(X)

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
y_test
```

Now get model parameters and see the effect to testing data.

```python
beta = result.params
beta
```

```python
def final_prob(x,beta):
  return 1/(1+np.exp(-np.dot(x,beta)))
y_score = final_prob(x_test,beta)
y_score
```

```python
y_series = pd.Series(y_score, index = y_test.index, name = 'model_score')
result_set = pd.concat([y_test,y_series],axis=1)
result_set['Bin_Pred'] = np.where(result_set['model_score'] > 0.5, 1, 0)
result_set['Bin_Pre_acc'] = np.where(result_set['survived'] == result_set['Bin_Pred'], 1, 0)
prediction_Rate = result_set['Bin_Pre_acc'].mean()
prediction_Rate
```
